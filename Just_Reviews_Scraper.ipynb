```python

import selenium
import pandas as pd
from selenium import webdriver

```

```python
df = pd.read_csv('/home/iii/MEGA/NYC Data/GoodReadsScraper/out.csv')
df
```

```python
colNames = ['book_review']

bookOutput = pd.DataFrame(columns=colNames)

bookOutput['book_title'] = df['book_title']
bookOutput['URL'] = df['goodreads']
```

```python
bookOutput.book_title[28]
```

```python
driver = webdriver.Chrome(executable_path='/home/iii/MEGA/NYC Data/GoodReadsScraper/chromedriver')


for bookIdx in bookOutput.index[1:35]:

    driver.get(bookOutput.URL[bookIdx])

    review_dict = {}


    reviews = driver.find_elements_by_xpath('//div[@class="friendReviews elementListBrown"]')


    for reviewIdx, review in enumerate(reviews):
#         print(bookOutput.book_title[reviewIdx])

        try:
            review.find_element_by_xpath('.//div[@class="reviewText stacked"]/span/a').click()
            text = review.find_element_by_xpath('.//div[@class="reviewText stacked"]/span/span[2]').text


        except:
#            print(f'No button for review {reviewIdx}!')
#             text = review.find_element_by_xpath('.//div[@class="reviewText stacked"]/span/span[1]').text   
            print(f'No button for review {idx}!')
            
            try:
                text = review.find_element_by_xpath('.//div[@class="reviewText stacked"]/span/span[1]').text   

            except Exception as e:
                print(type(e), e)
                text = "short review"   


        review_dict[reviewIdx] = text
        

        ###??? How to enter each dict into the proper row?
        bookOutput['book_review'][bookIdx] = review_dict
        
```
